```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.metrics import accuracy_score
from sklearn.model_selection import cross_val_score

from analyze_movie.notebook.demo_classifier import prediction_quality

df = pd.read_csv('../datas/titanic_clean.csv')
df.info()
```

```python
from sklearn.model_selection import train_test_split

# critères de base
x = df[['Pclass', 'Age', 'Sex_male', 'Embarked_C', 'Embarked_Q', 'Embarked_S']]

# Critère à prédire
y = df['Survived']

# splitter les données : 80% de training /20% de test
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
tree_1 = DecisionTreeClassifier()
tree_1.fit(x_train, y_train)
```

```python
import matplotlib.pyplot as plt

# 6. Affichage graphique de l'arbre pour débugger avec plot_tree
plt.figure(figsize=(10,6))
plot_tree(tree_1, feature_names=['Pclass', 'Age', 'Sex_male', 'Embarked_C', 'Embarked_Q', 'Embarked_S'], class_names=['Non', 'Oui'], filled=True)
plt.title("A survécu ?")
plt.show()
```

```python
from sklearn.tree import DecisionTreeClassifier, plot_tree

model = DecisionTreeClassifier(max_depth=5)
model.fit(x_train, y_train)
```

```python
import matplotlib.pyplot as plt

# 6. Affichage graphique de l'arbre
plt.figure(figsize=(10,6))
plot_tree(model, feature_names=['Pclass', 'Age', 'Sex_male', 'Embarked_C', 'Embarked_Q', 'Embarked_S'], class_names=['Non', 'Oui'], filled=True)
plt.title("A survécu ?")
plt.show()
```

```python
from sklearn.metrics import accuracy_score

prediction_quality = model.predict(x_test)

accuracy = accuracy_score(y_test, prediction_quality)

print(f"Accuracy: {accuracy}")
```
