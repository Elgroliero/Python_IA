```python
import pandas as pd
from skimage.metrics import mean_squared_error
from sklearn.model_selection import train_test_split

# 6 personnes
data = {
    'age' :             [22, 30, 25, 35, 19, 5],
    'like_music':       [1, 0, 0, 1, 1, 1],
    'friends_go_party': [1, 0, 1, 0, 1,0],
    'go_party':         [1, 0, 1, 1, 0, 0]
}
df = pd.DataFrame(data)
```

## Séparation des données

```python
x = df[['age', 'like_music', 'friends_go_party']]
y = df['go_party']

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
from sklearn.tree import DecisionTreeClassifier, plot_tree

model = DecisionTreeClassifier()
model.fit(x_train, y_train)
```

```python
# Exemple : prédire si cette personne va à la fête

# 1 : Créer une personne
new_person = pd.DataFrame([[28, 1, 0]], columns=['age', 'like_music', 'friends_go_party'])

#2 : Prédire
prediction = model.predict(new_person)
print(prediction)
```

```python
import matplotlib.pyplot as plt

# 6. Affichage graphique de l'arbre
plt.figure(figsize=(10,6))
plot_tree(model, feature_names=['age', 'like_music', 'friends_go_party'], class_names=['Non', 'Oui'], filled=True)
plt.title("Va à la fête ?")
plt.show()
```

```python
from sklearn.metrics import mean_squared_error

prediction_quality = model.predict(x_test)

mse = mean_squared_error(y_test, prediction_quality)

print(f"MSE : {mse}")
```

```python
from sklearn.metrics import accuracy_score

accuracy = accuracy_score(y_test, prediction_quality)

print(f"Accuracy : {accuracy}")
```
