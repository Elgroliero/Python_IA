```python
import pandas as pd
from sklearn.model_selection import train_test_split

#Prédire les notes que les élèves auront au BAC, selon les notes qu'ils ont eu dans les matières sur l'année
#Simuler données réelless
# Deux informations cruciales: les données sur lequelles s'appuyer pour prédire un résultat (notes en dehors du bac)
# Les données (les notes du bac)
df = pd.DataFrame({
    'maths' : [12, 8, 15, 10],
    'français' : [14, 9, 16, 11],
    'histoire' : [13, 10, 15, 14],
    'bac' : [15, 18, 9, 4],
})
```

```python
# Etape : Séparation
# x = données sur lesquelles s'appuyer
x = df[['maths', 'français', 'histoire']]

# y = la prédiction
y = df['bac']

# x_train = 80% des données réelles sur lesquelles s'appuyer pour s'entraîner (80% des notes de l'élève)
# x_test = 20% restants des données réelles sur "lesquelles s'appuer" pour qu'il s'auto-test (20% des notes de l'élève

# y_train = 80% des données réelles "prédites" pour s'entrainer
# y_test = 20% restants des données réelles "prédites" pour qu'il s'auto-test

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
# Etape : Dire à l'IA de s'entraîner

# Instancier un modèle vidéo pré écrit
model = LinearRegression()

# Entrainer le modèle à prédire y_train en s'appuyant sur x_train
# Entrainer le modèle à prédire les notes du BAC en s'appuyant sur les notes de l'année
model.fit(x_train, y_train)
```

```python
# Tester la prédiction

df_predictions = pd.DataFrame({
    'maths' : [5, 20],
    'français' : [4, 6],
    'histoire' : [14, 7],
})

#Tester de prédire les notes du BAC avec deux élèves
# 1. Utiliser le model.predict en parenthèses (les nouvelles données sur lesquelles s'appuyer pour prédire)
# 2. Utiliser le model.predict; le return (la prédiction)
#-----------------------
# 1. Utiliser le model.predict entre parenthèses (les notes des nouveaux élèves)
# 2. Pour prédire; le return de leurs notes au bac
predictions_test = model.predict(df_predictions)

print(predictions_test)
```

```python
# Regarder le MSE
from sklearn.metrics import mean_squared_error

# Pour avoir le MSE :
# Il faut le même nombre d'élèves que dans le y_test
# Il faut le même nombre d'élèves prédicts que dans les 20% des élèves réels (pourquoi 20% pcq on split à 0.2
# Dans notre cas on avait 4 élèves réels, 20% de 4 = 1
# Donc 1 élève
df_un_seul_eleve = pd.DataFrame({
    'maths' : [5],
    'français' : [4],
    'histoire' : [14],
})

predict_test = model.predict(df_un_seul_eleve)

mse = mean_squared_error(y_test, predict_test)
print(mse)
```

```python
# Exporter ton modèle en pkl ; pickle est l'outil qui sert à dumper des données
# Réutiliser ton modèle à l'extérieur (comme un exécutable)

import pickle

# Save
with open('../model/model_predic_bac.pkl', 'wb') as file:
    pickle.dump(model, file)

```
